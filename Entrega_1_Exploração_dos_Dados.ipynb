# Exploração dos Dados

## Justificativa da escolha das estatísticas e gráficos:

> Gráfico de Barra:
>> Este tipo de gráfico foi escolhido pois apresenta de forma clara a diferença entre a quantidade de amostras para cada classe em atributos categóricos.

> Histograma:
>> Este tipo de gráfico foi escolhido pois apresenta de forma clara a quantidade de amostras para N intervalos de um atributo contínuo.

> Estatísticas Descritivas:
>> Foram usadas medidas de tendência central e medidas de disperção. A média foi escolhida pois representa a concentração dos dados, mas também foram usados os valores de mediana (valor da amostra central dos dados ordenados) e moda (valor que mais se repete), pois assim temos um entendimento mais fiel dos dados, diminuindo a influência dos outliers. Também foram usados os percentil 25% e 75%, para complementar o entendimento de como os valores de um atributo estão distribuidos. Os valores mínimos e máximo foram utilizados para conhecer a amplitude dos valores do atributo. A variância e desvio padrão também foram calculadas para entender como está a dispersão dos valores.

> Boxplot: 
>> Basicamente uma representação visual das estatisticas apresentadas anteriormente, esse tipo de gráfico foi utilizado pois ele ajuda a avaliar a distribuição dos valores do atributo, também apresenta potenciais outliers dos dados.

## Funções Auxiliares

```python
def grafico_atributo(df, atributo, anotacao="", hue_atributo = 'sem_hue', size = (8,5)):
    plt.figure(figsize=size)

    if hue_atributo == 'sem_hue':
        ax = sns.countplot(df[atributo]);
    else: 
        ax = sns.countplot(x = df[atributo], hue=df[hue_atributo]);
    
    ax.set_title(f'Atributo {atributo} {anotacao}', fontsize=15)
    plt.show();

def histograma_atributo(df, atributo, anotacao, bins = 10, size = (8,5)):
    plt.figure(figsize=size)
    
    ax = sns.histplot(data=df[atributo], bins=bins);
    ax.set_title(f'Histograma {atributo} {anotacao}', fontsize=15)
    plt.show()

def estatisticas_descritivas(df, atributo, anotacoes):
    print(anotacoes)
    print(f'Média: {df[atributo].mean()}')
    for i in range(len(df[atributo].mode())):
        print(f'Moda: {df[atributo].mode()[i]}')
    print(f'Percentil (25%): {df[atributo].quantile(q=0.25)}')
    print(f'Percentil (50% ou Mediana): {df[atributo].quantile(q=0.5)}') #print(df[atributo].median())
    print(f'Percentil (75%): {df[atributo].quantile(q=0.75)}')
    print(f'Valor Máximo: {df[atributo].max()}')
    print(f'Valor Mínimo: {df[atributo].min()}')
    print(f'Variância: {df[atributo].var()}')
    print(f'Desvio Padrão: {df[atributo].std()}')
    print("\n\n")

def boxplot_atributo(df, atributo, anotacao, size = (8,5)):
    plt.figure(figsize=size)
    
    ax = sns.boxplot(df[atributo]);
    ax.set_title(f'Boxplot {atributo} {anotacao}', fontsize=15)
    plt.show()
```
